# Aligning with Hypertools

The align feature uses hyperalignemnt (Haxby et al, 2011) to align a list of high-dimensional arrays to a common space via linear transformations. It also provides the option to use the Shared Response Model (SRM) for alignment, if preferred, via the Brain Imaging Analysis Kit (brainiak).

Alignment can be particularly useful in exploring statistical properties and/or similarities of datasets that are not in the same coordinate system (such as fMRI data from visual areas of participants watching a movie, and the movie data itself).

## Import Hypertools

```python
import hypertools as hyp
```

## Load your data

In this case, we have used one of the sample datasets built into the package

```python
weights = hyp.load('weights_avg')
```

We can take a cursory look at the data structure (namely, the type and size of 'weights' and its components)

```python
print("Length: " + str(len(weights))); print("Type: " + str(type(weights))); print(); print('Each element contains a '+ str(type(weights[0][0]))+' consisting of '+ str(type(weights[0][0][0]))+'s')
```

## Visaulize raw data

First, we can see how the first hundred data points from two arrays in the weights data look when plotted together.

```python
geo_raw = hyp.plot([weights[0][0:200],weights[1][0:200]])
```

## Align data (hyperalign)

Next, we can align the two datasets and visualize the aligned data.

```python
aligned_data = hyp.align([weights[0][0:200],weights[1][0:200]])
```

```python
geo_aligned = hyp.plot(aligned_data)
```

# Align data (SRM)

You may also passs 'SRM' to the align argument, to use the Shared Response Model for alignment.

```python
aligned_SRM = hyp.align([weights[0][0:200],weights[1][0:200]], align='SRM')
```

```python
geo_SRM = hyp.plot(aligned_SRM)
```
