# Clustering with Hypertools

The cluster feature performs clustering analysis on the data (an arrray, dataframe, or list) and returns a list of cluster labels. 

The default clustering method is K-Means, with a variety of clustering schemes made available.

Note that, if a list is passed, the arrays will be stacked and clustering will be performed across *all* lists (not within each list).

## Import Hypertools

```python
import hypertools as hyp
```

## Load your data

In this case, we have used one of the sample datasets built into the package.

```python
mushrooms = hyp.load('mushrooms')
```

We can peek at the first few rows of the dataframe using the pandas feature 'head(n)'

```python
mushrooms.head(n=5)
```

## Obtain cluster labels

To obtain data labels, simply pass the data to hyp,cluster

```python
labels = hyp.cluster(mushrooms)
```

```python
labels
```

In this case, three labels were assigned (0, 1, and 2)

```python
set(labels)
```

## Specify number of cluster labels

You can also specify the number of desired clusters by passing an int to the n_clusters argument, as below.

```python
labels_5 = hyp.cluster(mushrooms, n_clusters = 5)
```

We can see that when we pass the int 5 to n_clusters, 5 cluster labels are assigned.

```python
set(labels_5)
```

## Different clustering models

```python
labels_agglom = hyp.cluster(mushrooms, cluster = 'AgglomerativeClustering')
```

```python
labels_agglom
```

We can see that the agglomerative clustering method assigns three feature labels (0, 1, and 2), as did the default clustering method (Kmeans) in our previous example.

```python
set(labels_agglom)
```

To obtain a desired number of feature labels, pass an int to the n_clusters argument.

```python
labels_agglom_5 = hyp.cluster(mushrooms, cluster = 'AgglomerativeClustering', n_clusters = 5)
```

We can see that when we pass the int 5 to n_clusters, 5 cluster labels are assigned.

```python
set(labels_agglom_5)
```
